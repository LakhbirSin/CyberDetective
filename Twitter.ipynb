```python
# importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data 
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('the Art of War since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('The song of Achille since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(' Batman : the Dark Night Return since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    

for i,tweet in enumerate(sntwitter.TwitterSearchScraper(' The Picture of Dorian Gray since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(' The Book Thief since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
tweets_df1
```

```python
#Remove special character
tweets_df1['Text'] =  tweets_df1['Text'].str.replace(r"[^a-zA-Z0-9]"," ")
tweets_df1
```

```python
# Save tweets table
tweets_df1.to_csv('Tweets.csv')
```
