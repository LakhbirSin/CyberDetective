```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
# Creating list to append tweet data 
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('The song of Achille since:2022-06-01 until:2023-01-01 ').get_items()): #declare a username 
    if i>1000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
```

```python
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
tweets_df1
```

```python
#Remove special character
tweets_df1['Text'] =  tweets_df1['Text'].str.replace(r"[^a-zA-Z0-9]"," ")
tweets_df1
```

```python
# Save tweets table
tweets_df1.to_csv('The_song_of_Achille.csv')
```

```python
Cloud = pd.read_csv("The_song_of_Achille.csv")
Cloud
```

```python
# Start with one review:
text = " ".join(i for i in Cloud['Text'])
# Stopword list
stopwords = set(STOPWORDS)
stopwords.update(['https','t','co','will','say','got','first','amp','don','go','s','de','m','je','La','re','d','c'])

# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
